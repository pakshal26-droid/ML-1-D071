**1. Use python libraries to build a decision tree classifier on Dataset 1. Analyze the results using confusion matrix and accuracy. Plot the Decision Tree.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix

df_iris = pd.read_csv("/Iris.csv")
X_iris = df_iris.iloc[:, :-1]
y_iris = df_iris.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

dt_classifier = DecisionTreeClassifier(criterion='gini', random_state=42)
dt_classifier.fit(X_train, y_train)

y_pred = dt_classifier.predict(X_test)

acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(f"Accuracy: {acc:.2f}")
print("Confusion Matrix:\n", cm)

plt.figure(figsize=(12, 8))
plot_tree(dt_classifier, filled=True, feature_names=X_iris.columns, class_names=df_iris.iloc[:, -1].unique())
plt.show()
```

**2. Write a code to show overfitting in the decision tree classifier built using Dataset 1. Use sklearn and matplotlib.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

df_iris = pd.read_csv("/Iris.csv")
X_iris = df_iris.iloc[:, :-1]
y_iris = df_iris.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

dt_overfit = DecisionTreeClassifier(max_depth=None, random_state=42)
dt_overfit.fit(X_train, y_train)

y_train_pred = dt_overfit.predict(X_train)
train_acc = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy (Overfit Model): {train_acc:.2f}")

y_test_pred = dt_overfit.predict(X_test)
test_acc = accuracy_score(y_test, y_test_pred)
print(f"Testing Accuracy (Overfit Model): {test_acc:.2f}")
```

**3. Implement Decision tree regressor on Dataset 2.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.preprocessing import LabelEncoder

df_car = pd.read_csv("/car_prediction_data.csv")

for col in df_car.columns:
    if df_car[col].dtype == 'object':
        le = LabelEncoder()
        df_car[col] = le.fit_transform(df_car[col])

X_car = df_car.iloc[:, :-1]
y_car = df_car.iloc[:, -1]

X_train_car, X_test_car, y_train_car, y_test_car = train_test_split(X_car, y_car, test_size=0.2, random_state=42)

dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train_car, y_train_car)

y_pred_car = dt_regressor.predict(X_test_car)

plt.figure(figsize=(12, 8))
plot_tree(dt_regressor, filled=True, feature_names=X_car.columns)
plt.show()
```
